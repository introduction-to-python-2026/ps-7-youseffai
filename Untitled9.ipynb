```python

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml
import pandas as pd


data=fetch_openml(name='autoMpg', version=3, as_frame=True)
```

```python
print(data.DESCR)
```

```python
df = data.frame

df.sample(9)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
features = df.columns
print(features)
selected_features =['Acceleration','Horse_power']
```

```python
features = list(df.columns)
print("Available features:", features)

print("Selected features: ", selected_features)
```

```python
fig, axs  = plt.subplots(1, len(selected_features), figsize = (16,3))

for ax, f in zip(axs, selected_features):
    ax.hist(df[f], bins=30, color='red', edgecolor='black')
    ax.set_xlabel(f)
```

```python
reference_feature = "Acceleration"
y = df[reference_feature]

fig, axs  = plt.subplots(1, len(selected_features), figsize = (20,5))

for ax, f in zip(axs, selected_features):
  ax.scatter(df[f], y)
  ax.set_xlabel(f)

plt.show()
```

```python

reference_feature = selected_features[1]
comparison_feature = selected_features[0]

plt.figure(figsize=(8, 6))
plt.scatter(df[reference_feature], df[comparison_feature], alpha=0.6)
plt.xlabel(reference_feature)
plt.ylabel(comparison_feature)


plt.savefig('correlation_plot.png')

plt.show()
```
